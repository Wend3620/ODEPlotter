```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
import numpy as np

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

width = 500
height = 550
t_mergin = 50

fig = go.Figure()
# Create streamline figure
for step in np.arange(-5, 5, 1):
    Y, X = np.meshgrid(x, y)
    #Actual function
    u = -1 - X**2 + Y
    v = 1 + step*X - Y**2 

    fig.add_trace(ff.create_streamline(x, y, u, v, arrow_scale=.1, visible = False, line_color = 'Teal').data[0])
# Create and add slider

fig.data[5].visible = True

steps = []

for i in range(len(fig.data)):
    step = dict(
        method="update",
        args=[{"visible": [False] * len(fig.data)},
              {"title": "Slider switched to step: " + str(i)}],  # layout attribute
    )
    step["args"][0]["visible"][i] = True  # Toggle i'th trace to "visible"
    steps.append(step)

sliders = [dict(
    active=5,
    currentvalue={"prefix": "Value of a: "},
    pad={"t": 20, 'b': 20},
    steps=steps,
    font = {'size' : 12, 'color': 'navy'}
)]

fig.update_layout(
    width = width,
    height = height,
    margin=dict(l=20, r=40, t=t_mergin, b=5),
    #paper_bgcolor="LightSteelBlue",
    sliders=sliders,
    title = dict(text = "Slider switched to step:", x = 0.03, y = (height - t_mergin+25)/height),
    annotations = [
        #  {
        #     'x': 0.5,
        #     'y': -0.15,
        #     'xref': 'paper',
        #     'yref': 'paper',
        #     'text': 'Footer',      # Add a footer
        #     'showarrow': False,
        #     'font': {'size': 10},
        # }
    ]
)

fig.show()
```
